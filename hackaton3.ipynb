```python
# Cambio el nombre de los campos Latitud y Longitud en el Dataframe acc2 por Latitud_WGS84 y Longitud_WGS84,
# como está en la tabla acc2023.

acc2 = acc2.rename(columns = {'Longitud':'Longitud_WGS84', 'Latitud':'Latitud_WGS84'})
acc2
```

```python
accidentes = pd.concat([acc2, acc2023], ignore_index=True)
```

```python
accidentes.info()
```

```python
# miro los campos que tienen valores nulos

nulos = accidentes.isnull().sum()
nulos
```

```python
# resumen de las filas que tienen filas con nulos
filas_con_nulos = accidentes[accidentes.isnull().any(axis=1)]
filas_con_nulos
```

```python
# reemplazar los valores nulos por Desconegut

accidentes.fillna("Desconegut", inplace=True)
```

```python
# vuelvo a mirar que ahora no quedan nulos

nulos = accidentes.isnull().sum()
nulos
```

```python
# eliminar registros duplicados

conteo_duplicados = accidentes.duplicated().sum()
conteo_duplicados
```

```python
# vamos a mirar que tipo de datos es cada columna
accidentes.info()
```

```python
accidentes.dtypes
```

```python
# vamos a cambiar a tipo de dato Longitud_WGS84 y Latitud_WGS84 a float
#accidentes['Longitud_WGS84'] = accidentes['Longitud_WGS84'].astype('float')

# No me deja cambiar a float porque tengo registros con valor: Desconegut.
# Voy a cambiar desconegut por cero.

accidentes['Longitud_WGS84'] = accidentes['Longitud_WGS84'].replace('Desconegut', 0)
#  ahora si. voy a cambiar a tipo de dato Longitud_WGS84 y Latitud_WGS84 a float
accidentes['Longitud_WGS84'] = accidentes['Longitud_WGS84'].astype('float')
```

```python
# hago lo mismo con la Latitud_WGS84
accidentes['Latitud_WGS84'] = accidentes['Latitud_WGS84'].replace('Desconegut', 0)
accidentes['Latitud_WGS84'] = accidentes['Latitud_WGS84'].astype('float')
```

## GRAFICOS

```python
accidentes['NK_Any'] = accidentes['NK_Any'].astype(str)
plt.figure(figsize=(8, 5))  
g = sns.countplot(x='NK_Any', data=accidentes, width = 0.5, color = '#4CAFE0') 
plt.bar_label(g.containers[0], fontsize = 10)
plt.title('Número total de accidents por any')
plt.xlabel(None)
plt.ylabel('Accidents')
plt.ylim(6000, None)
plt.show()
```

En este gráfico, se representa el número de accidentes por año. Se aprecia claramente que en el 2020, una grna disminución de accidentes debido al confinamiento de la pandemia seguramente. 
También vemos que el año 2019 es el que tubo el mayor número de accidentes. Por suerte después de la pandemia el número de accidentes se ha redudido en una media aproximada de 2000.

```python
plt.figure(figsize=(9, 4))  
g = sns.countplot(x='Nom_districte', data=accidentes, width = 0.5, color= '#4CAFE0') 
plt.bar_label(g.containers[0], fontsize = 10)
plt.title('Número total de accidents per districte')
plt.xlabel(None)
plt.ylabel('Num_accidents')
plt.xticks(rotation=80)
plt.show()
```

Seguido por Sant Martí con 5615 y Sants- Montjuïc con 5902. No es despreciable temapoco el número de accidentes de Sarrià-Sant Gervasi con 4902.

Es muy curioso, todo y que son valores similares, la diferencia máxima es de un 7% aproximadamente, a medida que avanza la semana los accidentes van en aumento.

```python
accidentes_selec = accidentes [['Nom_districte','Descripcio_causa_mediata','Numero_expedient']]
accidentes_selec
```

```python
accidentes['Descripcio_causa_mediata'].unique()
```

```python
accidentes_selec2 = accidentes_selec.groupby(['Descripcio_causa_mediata', 'Nom_districte'])['Numero_expedient'].count().reset_index()
accidentes_selec2
accidentes_selec2 = accidentes_selec2.rename(columns = {'Numero_expedient': 'Num_acc'})
```

```python
accidentes_selec2
```

```python
cuenteo = accidentes_selec2.pivot_table(index='Descripcio_causa_mediata', 
                                columns='Nom_districte', 
                                values='Num_acc', 
                                aggfunc='sum', 
                                fill_value=0)
```

```python
plt.figure(figsize=(6, 6))  # Ajusta el tamaño del gráfico
sns.heatmap(cuenteo, annot=True, cmap="Blues", linewidths=.5, fmt="d",annot_kws={"size": 7})
plt.xticks(rotation=85)
plt.tick_params(axis='both', which='major', labelsize=7)
plt.show()
```

Por último, representamos el número de accidentes según la causa por distrito. Existe una gran variedad de causas por las cuales se producen accidentes. Las principales causas son: la manca d'atenció a la conducció y gir indegut o sense precaució. La más importante es la primera comentada. Otra causa considerable es NO repectar les distàncies.
